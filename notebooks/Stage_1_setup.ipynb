# Stage 1 Setup

This notebook creates the initial local music library from `links.csv` using `src/download_youtube_audio.py`, then inspects the resulting registry.

```python
from pathlib import Path
import sys

import pandas as pd

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / 'links.csv').exists() and (PROJECT_ROOT.parent / 'links.csv').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
SRC_DIR = PROJECT_ROOT / 'src'
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

import download_youtube_audio as dya
```

## Source Library

Preview the track list that seeds the local library.

```python
links = pd.read_csv(PROJECT_ROOT / 'links.csv')
links.head()
```

## Build Registry

Run the downloader to create the local audio files and registry.

```python
rows = dya.read_rows(dya.INPUT_CSV)
summary = dya.process_rows(rows)
```

## Successful Downloads

Inspect the tracks that were newly downloaded in this run.

```python
successful_downloads = pd.DataFrame(summary['downloaded_items'])
successful_downloads.head()
```

## Registry Preview

Check the top of `music_registry.csv` after the download step.
